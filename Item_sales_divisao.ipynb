```python
import pandas as pd
import numpy as np
```

O item_sales foi dividido em 7 partes para possibilitar a manipulação

```python
item_sales_parte_1 = pd.read_csv('.\datasets\item_sales_parte_1.csv')
```

considerando que o id já possui o item, category, department  e o store_code podemos apagar essas colunas.

```python
item_sales_parte_1 = item_sales_parte_1.drop(['item', 'category', 'department','store_code'], axis = 1)
```

```python
item_sales_parte_1.head()
```

Passo para transformar a notação de dias na data correta

```python
dia_colunas = [col for col in item_sales_parte_1.columns if col.startswith('d_')]
```

```python
start_date = pd.to_datetime('2011-01-29')
```

```python
dates = [start_date + pd.DateOffset(days=int(col.split('_')[1]) - 1) for col in dia_colunas]
```

```python
dates
```

```python
item_sales_parte_1.rename(columns=dict(zip(dia_colunas, dates)), inplace=True)
```

```python
item_sales_parte_1
```

Mais uma divisão do df para possibilitar a manipulação

Aqui vamos aplicar o melt para transformar as colunas das datas em linhas

```python
item_sales_parte_1 = item_sales_parte_1.melt(id_vars=['id', 'store', 'region'], 
                    value_vars=dates, 
                    var_name='Data', 
                    value_name='Valor')
```

```python
item_sales_parte_1
```

Vamos agrupar os dados em semanas e meses

```python
item_sales_parte_1_weekly = item_sales_parte_1.groupby([pd.Grouper(key='Data', freq='W'),'id','store']).sum()
```

```python
item_sales_parte_1_weekly.head(14)
```

```python
item_sales_parte_1_monthly = item_sales_parte_1.groupby([pd.Grouper(key='Data', freq='M'),'id','store']).sum()
```

```python
item_sales_parte_1_monthly
```

salvando os df em csv

```python
item_sales_parte_1_monthly.to_csv('item_sales_parte_1_monthly.csv')
```

```python
item_sales_parte_1_weekly.to_csv('item_sales_parte_1_weekly.csv')
```
